```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("Customer-Call-List.xlsx")
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.Phone_Number = df.Phone_Number.str.replace( '[^a-zA-Z0-9]' , '',  regex=True) 

#usei para substituir todos os caracteres que NÃO sejam letras maiúsculas, letras minúsculas ou dígitos.
```

```python
df
```

```python
df.Phone_Number = df.Phone_Number.str.replace('[a-zA-Z]', '', regex = True) #remoção das letras
```

```python
df.Phone_Number = df.Phone_Number = df.Phone_Number.str.strip() #para retirar os espaços 
```

```python
df
```

```python
df.Phone_Number = df.Phone_Number.replace('' , np.nan) 
#preenchi as celulas vazias por NaN, para facilitar a remoção das linhas caso seja necessario!
```

##### columns = 'Last_Name'

```python
df[['Last_Name']]
```

```python
df.Last_Name = df.Last_Name.str.replace('[^a-zA-Z]', '', regex = True) 
#usei a mesma logica para retirar metacaracteres da coluna lastname


#df.Last_Name.str.rstrip('_')
#df.Last_Name.str.lstrip('123._/') poderia ser retirados dessa maneira, porém, seria bem mais demorado!
```

```python
df.Last_Name = df.Last_Name.str.strip() #remoção de espaços desnecessarios 

```
