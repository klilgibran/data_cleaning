```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("Customer-Call-List.xlsx")
```

```python
df.columns
```

```python
df.columns = ['ID_cliente' , 'Nome', 'Sobrenome' ,'Contatos', 'Endereço',  'Cliente_Pagante', 'Não_Contatar' , 'Not_Useful_Column']
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

#### - iniciei pela coluna de contatos, por considerar a mais trabalhosa! 

```python
#para substituir todos os caracteres que NÃO sejam letras maiúsculas, letras minúsculas ou metacacacteres.
df.Contatos = df.Contatos.str.replace( '[^a-zA-Z0-9]' , '',  regex=True) 

 
```

```python
df
```

```python
df.Contatos = df.Contatos.str.replace('[a-zA-Z]', '', regex = True) #remoção das letras
```

```python
df.Contatos 
```

```python
df.Contatos = df.Contatos = df.Contatos.str.strip() #retirar os espaços desnecessarios 
```

```python
df.Contatos = df.Contatos.replace('' , np.nan) 
#preenchi as celulas vazias por NaN, para facilitar a remoção das linhas caso seja necessario!
```

```python
 #havia valores do tipo float, converti todos para strings, para aplicar outro método. 
df.Contatos = df.Contatos.astype(str)
```

```python
df.Contatos = df['Contatos'].apply(lambda x: ('('+ x[0:2] + ')' + x[2:6] + '-' + x[6:10] ))
```

```python
df.head()
```

```python
#substituir os espaços com "letras" por NaN, para facilitar a manipulação!
df['Contatos'] = df['Contatos'].replace("(na)n-", np.nan )   
```
